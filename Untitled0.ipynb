```python
# Assuming 'train_main_loan.csv' is the uploaded file
train = pd.read_csv('train_main_loan.csv')
train['ACTUAL_ROI'].fillna(train['ACTUAL_ROI'].mean(), inplace=True)
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import numpy as np
data = train.drop(['ID', 'DATE_OF_BIRTH', 'Reported_Date', 'DATE_REPORTED_AND_CERTIFIED', 'DATE_OF_LAST_PAYMENT'], axis=1)
x = pd.get_dummies(data, columns=['GENDER', 'ACCOUNT_TYPE', 'OWNERSHIP_TYPE', 'OCCUPATION_TYPE'])
# Handle missing values and outliers
x['EMI_AMOUNT'].fillna(train['EMI_AMOUNT'].mean(), inplace=True)
x['REPAYMENT_TENURE'].fillna(train['REPAYMENT_TENURE'].mean(), inplace=True)
x['ACTUAL_PAYMT_AMT'].fillna(train['ACTUAL_PAYMT_AMT'].mean(), inplace=True)
x['AMOUNT_OVERDUE'].fillna(train['AMOUNT_OVERDUE'].mean(), inplace=True)
x['COLLATERALVALUE'].fillna(train['COLLATERALVALUE'].mean(), inplace=True)
x['PAYMENT_HISTORY_1'].fillna('0', inplace=True)
x['PAYMENT_HISTORY_2'].fillna('0', inplace=True)

# Define a function to process each entry in the column
def process_payment_history(entry):
    if not pd.api.types.is_numeric_dtype(entry):
        entry = entry.replace('XXX', '000')
        entry = entry.replace('STD', '000')
        entry = entry.replace('SMA', '060')
        entry = entry.replace('SUB', '090')
        entry = entry.replace('DBT', '090')
        entry = entry.replace('LSS', '090')

        entry = ''.join(['000' if c not in '0123456789' else c for c in entry])

    return pd.to_numeric(entry, errors='coerce')

# Apply the function to the entire column
x['PAYMENT_HISTORY_1'] = -x['PAYMENT_HISTORY_1'].apply(process_payment_history).map(int)
x['PAYMENT_HISTORY_2'] = -x['PAYMENT_HISTORY_2'].apply(process_payment_history).map(int)

# Scaling and Normalizing very large integer values
scaler = StandardScaler()
x[['PAYMENT_HISTORY_1','PAYMENT_HISTORY_2']] = scaler.fit_transform(x[['PAYMENT_HISTORY_1', 'PAYMENT_HISTORY_2']])
print(len(x))
# Convert date columns to datetime type
x['PAYMENT_HISTORY_START_DATE'] = pd.to_datetime(x['PAYMENT_HISTORY_START_DATE'], format='%d-%m-%Y')
x['PAYMENT_HISTORY_END_DATE'] = pd.to_datetime(x['PAYMENT_HISTORY_END_DATE'], format='%d-%m-%Y')
x['DATE_OPENED'] = pd.to_datetime(x['DATE_OPENED'], format='%d-%m-%Y')
reference_date = pd.to_datetime('01-01-2024', format='%d-%m-%Y')

# Calculate payment duration and create new columns
x['PAYMENT_DURATION'] = -(x['PAYMENT_HISTORY_END_DATE'] - x['PAYMENT_HISTORY_START_DATE']).dt.days
x['ACCOUNT_OPENING_DURATION'] = -(x['DATE_OPENED'] - reference_date).dt.days

# Remove less important columns
x = x.drop(['LOAN_CLASSIFICATION','DATE_OPENED', 'PAYMENT_HISTORY_START_DATE', 'PAYMENT_HISTORY_END_DATE', 'OCCUPATION_TYPE_R18', 'GENDER_Other'], axis=1)

# Handle infinite or NaN values
x.replace([np.inf, -np.inf], np.nan, inplace=True)
x.fillna(0, inplace=True)

# Split the data into training and testing sets

# Split the data into training and testing sets
scaler = StandardScaler()
scaled_features = scaler.fit_transform(x.drop('ACTUAL_ROI', axis=1))  # Exclude 'ACTUAL_ROI' from scaling
x_scaled = pd.DataFrame(scaled_features, columns=x.columns[:-1])  # Exclude the last column 'ACTUAL_ROI'
x_scaled['ACTUAL_ROI'] = x['ACTUAL_ROI']  # Add 'ACTUAL_ROI' back to the DataFrame



# Assuming 'x' is your DataFrame containing the data, and 'ACTUAL_ROI' is the target column
# Replace 'x' and 'ACTUAL_ROI' with your actual DataFrame and column names

# Set the number of rows and columns dynamically based on the number of features
num_features = len(x_scaled.columns) - 1  # Exclude 'ACTUAL_ROI'
num_rows = num_features // 3 + 1
num_cols = min(3, num_features)

# Set the size of the overall plot
plt.figure(figsize=(15, 5 * num_rows))

# Loop through each feature column and create a scatter plot
for i, feature in enumerate(x_scaled.columns[:-1], 1):  # Exclude the last column 'ACTUAL_ROI'
    plt.subplot(num_rows, num_cols, i)
    plt.scatter(x_scaled[feature], x_scaled['ACTUAL_ROI'], alpha=0.5)
    plt.title(f'Scatter Plot: {feature} vs ACTUAL_ROI')
    plt.xlabel(feature)
    plt.ylabel('ACTUAL_ROI')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
